#  Importing libararies

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from matplotlib.colors import ListedColormap
from sklearn.metrics import confusion_matrix 
from sklearn import metrics
import seaborn as sns
import warnings;
warnings.filterwarnings('ignore');
```

```python
# importing dataset
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
dataset
```

# Extracting variables

```python
X = dataset.iloc[:,[2,3]].values
X
```

```python
Y = dataset.iloc[:,4].values
Y
```

#   Heatmap to see correlation

```python
sns.heatmap(dataset.corr(numeric_only=True), annot=True, cmap='RdYlGn')
```

# Train test split

```python
X_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=1/3,random_state=32)
```

# Feature scaling

```python
standard_Scaler = StandardScaler()
X_train = standard_Scaler.fit_transform(X_train)
x_test = standard_Scaler.transform(x_test)
```

```python
x_test
```

# instantiating and fitting the model to training dataset

```python
log_reg = LogisticRegression(random_state=0)
log_reg.fit(X_train,y_train)
```

# Prediction for test dataset

```python
y_pred = log_reg.predict(x_test)
y_pred
```

```python
print("predicted values",y_pred)
print("actual values", y_test)
```

# visualising the training set result 

```python
X_set, y_set = X_train, y_train
X1, X2 = np.meshgrid(
    np.arange(start=X_set[:, 0].min() - 1, stop=X_set[:, 0].max() + 1, step=0.01),
    np.arange(start=X_set[:, 1].min() - 1, stop=X_set[:, 1].max() + 1, step=0.01)
)
plt.contourf(
    X1, X2, 
    log_reg.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
    alpha=0.75, cmap=ListedColormap(('red', 'green'))
)
plt.xlim(X1.min(),X1.max())
plt.ylim(X2.min(),X2.max())

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c=ListedColormap(('yellow', 'blue'))(i), label=j)

plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

# visualising the testing result

```python
X_set,y_set = x_test,y_test
X1,X2 = np.meshgrid(
    np.arange(start=X_set[:, 0].min() - 1, stop=X_set[:, 0].max() + 1, step=0.01),
    np.arange(start=X_set[:, 1].min() - 1, stop=X_set[:, 1].max() + 1, step=0.01)
)

plt.contourf(
    X1, X2, 
    log_reg.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
    alpha=0.75, cmap=ListedColormap(('black', 'yellow'))
)
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c=ListedColormap(('red', 'green'))(i), label=j)

plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

# confusion matrix

```python
from sklearn.metrics import confusion_matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print(conf_matrix)
```

```python
import seaborn as sns
fig, ax = plt.subplots(figsize = (6,5))
sns.heatmap(conf_matrix,annot = True,ax=ax)
```

# Accuracy

```python
# accuracy
print("Accuracy : ", metrics.accuracy_score(y_test,y_pred))
```

```python
# precision
print("Precision : ", metrics.precision_score(y_test,y_pred))
```

```python
#recall
print("Recall : ", metrics.recall_score(y_test,y_pred))
```

# ROC and AUC

```python
y_pred_proba = log_reg.predict_proba(x_test)[:,1]
fpr, tpr, _ = metrics.roc_curve(y_test, y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)
plt.figure(figsize =(10,12))
plt.plot(fpr,tpr,label = "auc="+str(auc))
plt.legend(loc=4)
plt.title("ROC")
plt.xlabel("FPR ----->")
plt.ylabel("TPR ------>")
plt.show()
```
