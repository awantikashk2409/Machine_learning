```python
# encolding


# encolding values
df.replace({'Fuel_Type':{'Petrol':0, 'Diesel':1, 'CNG':2}}, inplace=True)



df.replace({'Transmission':{'Manual':0, 'Automatic':1}}, inplace=True)
```

```python
df.head()
```

```python
df.replace({'Seller_Type':{'Dealer':0, 'Individual':1}}, inplace=True)
```

```python
df.head()
```

```python
# Drop unwanted column and target column
X = df.drop(['Car_Name', 'Selling_Price'], axis=1)

# save target column in to Y
y = df['Selling_Price']
```

### Linear Regression

```python
# Train test split 

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# Prediction

pred = lr.predict(X_train)
```

```python
# Error 
from sklearn.metrics import r2_score
rsq = r2_score(y_train, pred)
print("R square Error: ", rsq)
```

```python
# Scatter plot 'Actual price' vs 'Predicted price'
plt.scatter(y_train, pred)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
```

```python
# plotting the actual and predicted value

c = [i for i in range(1, len(y_train)+1, 1)]
plt.plot(c, y_train, color='r', linestyle='-', label='Actual Values')
plt.plot(c, pred, color='g', linestyle='-', label='Predicted Values')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Prediction Using X_train')
plt.legend()
plt.show()
```

```python

pred_2 = lr.predict(X_test)
rsq = r2_score(y_test, pred_2)
print('R Square Error y_test: ',rsq)
```

```python
c = [i for i in range (1,len(y_test)+1,1)]
plt.plot(c,y_test,color='r',linestyle='-',label="Actual Values")
plt.plot(c,pred_2,color='g',linestyle='-',label="Predicted Values")
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Prediction using x_test')
plt.legend()
plt.show()
```
